```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)

lado_kohonen = [3,5,7,10,20,30,40,50]
n_clases = [9,24,43,63,71,83,88,91]
error_cuantificable = [0.32,0.2,0.13,0.10,0.09,0.08,0.07,0.07]
error_topologico = [0.03,0.04,0.01,0.03,0.08,0.04,0.1,0.15]

plt.xlabel('Lado kohonen')
plt.ylabel('Clases')
# Influencia del lado KHONONEN en número de clases
linea_clases = plt.plot(lado_kohonen,n_clases,'r')
plt.twinx()
# Influencia del lado KHONONEN en errores
plt.ylabel('Errores')
# Error cuantificable
linea_cuantificable = plt.plot(lado_kohonen,error_cuantificable,'g') 
plt.twiny()
# Error topológico
linea_topologico = plt.plot(lado_kohonen,error_topologico,'b') 

# EL MEJOR LADO DE KOHONEN ES 30
```
